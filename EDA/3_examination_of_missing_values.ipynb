# Eksik Değerlerin İncelenmesi

```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

```python
# veri setimizde hiç eksik gözlem(değer) var mıdır?
# any() : Bir tane gözlemimizde bile eksik değer var ise true döndürür.(or işlemi)
df.isnull().values.any()
```

```python
# hangi değişkende kaç tane eksik gözlem var?
df.isnull().sum()
```
