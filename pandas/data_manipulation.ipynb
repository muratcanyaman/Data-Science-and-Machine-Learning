## İleri Toplulaştırma İşlemleri(aggregate, filter, transform, apply)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
#aggregate
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min", np.median, max])
```

```python
df.groupby("gruplar").aggregate({"degisken1": "min", "degisken2": "max"})
```

```python
#filter
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
def filter_func(x):
    return x["degisken1"].std() > 9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

```python
#transform
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
df_a = df.iloc[:,1:3]
```

```python
df_a.transform(lambda x : x-x.mean())
```

```python
df_a.transform(lambda x : (x-x.mean()) / x.std())
```

```python
#transform fonksiyonu vektörizasyon yaparak çalışır.Yani Değişken(sutün) bazında çalışır.
#Yukarıdaki lambda fonksiyonunda x olarak gösterilen aslında degisken1 ve degisken2 sutünlarıdır.
```

```python
#apply
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
df.groupby("gruplar").apply(np.sum, axis = 0)
```

```python
df.groupby("gruplar").apply(np.mean)
```

# Pivot Tablolar

```python
# not : Pivot tablolar groupby'ın çok boyutlu versiyonudur.
```

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# cinsiyete göre hayatta kalanların ortalamasını inceleyelim.
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
# groupby aracılığıyla pivot işlemi
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean").unstack()
```

```python
# pivot ile table
```

```python
titanic.pivot_table("survived", index = "sex", columns = "class", observed = False)
```

```python
age = pd.cut(titanic["age"], [0, 18, 90])
age.head(10)
```

```python
titanic.pivot_table("survived", index = ["sex", age], columns = "class", observed = False)
```
