## Gruplama İşlemleri(Grouping)

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Gruplar" : ['A', 'B', 'C', 'A', 'B', 'C'],
                   "Veri" : [10,11,52,23,43,55]}, columns = ["Gruplar", "Veri"])
df
```

```python
df.groupby("Gruplar")
```

```python
df.groupby("Gruplar").mean()
```

```python
df.groupby("Gruplar").sum()
```

## İleri Toplulaştırma İşlemleri(aggregate, filter, transform, apply)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
#aggregate
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min", np.median, max])
```

```python
df.groupby("gruplar").aggregate({"degisken1": "min", "degisken2": "max"})
```

```python
#filter
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
def filter_func(x):
    return x["degisken1"].std() > 9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

```python
#transform
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
df_a = df.iloc[:,1:3]
```

```python
df_a.transform(lambda x : x-x.mean())
```

```python
df_a.transform(lambda x : (x-x.mean()) / x.std())
```

```python
#transform fonksiyonu vektörizasyon yaparak çalışır.Yani Değişken(sutün) bazında çalışır.
#Yukarıdaki lambda fonksiyonunda x olarak gösterilen aslında degisken1 ve degisken2 sutünlarıdır.
```

```python
#apply
```

```python
df = pd.DataFrame({'gruplar':  ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar','degisken1','degisken2'])
df
```

```python
df.groupby("gruplar").apply(np.sum, axis = 0)
```

```python
df.groupby("gruplar").apply(np.mean)
```
