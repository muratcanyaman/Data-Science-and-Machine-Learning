# Birleştirme(Join) İşlemleri

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns = ["var1","var2", "var3"])
df1 
```

```python
df2 = df1 + 99
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
?pd.concat
```

```python
pd.concat([df1,df2], ignore_index = True)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df2.columns = ["var1", "var2", "deg3"]
```

```python
df2
```

```python
df1
```

```python
pd.concat([df1,df2])
```

```python
#inner join:Bildiğimiz kümelerde kesişim işlemi ile aynı mantıkta çalışır.Ortak olan sütunları birleştirir.
pd.concat([df1,df2], join = "inner",ignore_index = True)
```

```python
pd.concat([df1,df2], join_axes = [df1.columns])
```

```python
# pandas'ın yeni sürümünde concat metodu artık 'join_axes' argümanını barındırmadığı için bu işlem hatalı sonlandı.
```
