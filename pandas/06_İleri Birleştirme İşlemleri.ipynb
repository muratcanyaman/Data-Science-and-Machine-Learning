```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'calisanlar' : ['Ali', 'Veli', 'Ayse', 'Fatma'],
                   'grup' : ['Muhasebe', 'Mühendislik', 'Mühendislik', 'İK']})
df1
```

```python
df2 = pd.DataFrame({'calisanlar' : ['Ayse', 'Ali', 'Veli', 'Fatma'],
                    'ilk_giris' : [2010, 2009, 2014, 2019]})
df2
```

```python
pd.merge(df1,df2)
```

```python
#alternatif
```

```python
pd.merge(df1,df2,on = 'calisanlar')
```

```python
#çoktan teke (many-to-one)
```

```python
df3 = pd.merge(df1,df2)
```

```python
df3
```

```python
df4 = pd.DataFrame({'grup' : ['Muhasebe', 'Mühendislik', 'İK'],
                    'mudur' : ['Caner', 'Mustafa', 'Berkcan']})
df4
```

```python
pd.merge(df3,df4)
```

```python
#Çoktan çoka(many-to-many)
```

```python
df5 = pd.DataFrame({'grup' : ['Muhasebe', 'Muhasebe', 'Mühendislik', 'Mühendislik', 'İK', 'İK'],
                    'yetenekler' : ['matematik', 'excel', 'kodlama', 'linux', 'excel', 'yonetim']})
df5
```

```python
df1
```
