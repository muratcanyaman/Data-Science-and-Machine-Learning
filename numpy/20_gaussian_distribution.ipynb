# Normal(Gauss) Daƒüƒ±lƒ±m üöÄ

‚ÑπÔ∏è *Normal Daƒüƒ±lƒ±m en √∂nemli daƒüƒ±lƒ±mlardan biridir.*<br>
‚ÑπÔ∏è *Alman Matematik√ßi Carl Friedrich Gauss'tan sonra 'Gauss Daƒüƒ±lƒ±mƒ±' olarak anƒ±lmaya ba≈ülanmƒ±≈ütƒ±r.*<br>
‚ÑπÔ∏è *Bir√ßok olayƒ±n olasƒ±lƒ±k daƒüƒ±lƒ±mlarƒ±na uyar.√ñrneƒüin IQ skoru, kalp atƒ±≈üƒ±...*<br>
‚ÑπÔ∏è *Normal Veri Daƒüƒ±lƒ±mƒ±na eri≈ümek i√ßin **random.normal()** metodu kullanƒ±lƒ±r.*<br>
‚ÑπÔ∏è *Bu metodun 3 parametresi var:*<br>
    ***loc()**:(mean) √ßanƒ±n tepe noktasƒ±nƒ±n bulunduƒüu yeri ifade eder.*<br>
    ***scale()**:(standart deviation) grafik daƒüƒ±lƒ±mƒ±nƒ±n ne kadar d√ºz olmasƒ± gerektiƒüini s√∂yler.*<br>
    ***size()**: geri d√∂nd√ºr√ºlen dizinin boyutu.*

*2x3'l√ºk rastgele normal daƒüƒ±lƒ±m √ºretelim:*

```python
from numpy import random
x = random.normal(size = (2, 3))
print(x)
```

*2x3'l√ºk,ortalamasƒ± 1,standart sapmasƒ± 2 olan rastgele normal daƒüƒ±lƒ±m √ºretelim:*

```python
from numpy import random
x = random.normal(loc = 1, scale = 2, size = (2, 3))
print(x)
```

## Normal Daƒüƒ±lƒ±mƒ±n G√∂rselle≈ütirilmesi üß†

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.normal(size = 1000), kind = "kde")
plt.show()
```
