# Logistic Distribution (Lojistik Daƒüƒ±lƒ±m) üöÄ

‚ÑπÔ∏è *Lojistik Daƒüƒ±lƒ±m b√ºy√ºmeyi ifade etmek i√ßin kullanƒ±lƒ±r.*<br>
‚ÑπÔ∏è *Lojistik Regresyon,Sinir Aƒülarƒ± gibi alanlarda makine √∂ƒürenmesinde yaygƒ±n olarak kullanƒ±lƒ±r.*<br>

‚ÑπÔ∏è *3 parametreye sahiptir:<br>
    **loc**: Zirvenin nerede olduƒüunu ifade eder.Varsayƒ±lan olarak 0'dƒ±r.*<br>
    **scale**: Standart Sapma, daƒüƒ±lƒ±mƒ±n d√ºzl√ºƒü√ºn√º ifade eder. Varsayƒ±lan olarak 1'dir.*<br>
    **size**: Geriye d√∂nd√ºr√ºlecek olan dizinin boyutunu ifade eder.*<br>

## √ñrnek

*Ortalamasƒ± 1 ve standart daƒüƒ±lƒ±mƒ± 2.0 olan bir lojistik daƒüƒ±lƒ±mdan 2x3'l√ºk √∂rnekler √ßekelim:*<br>

```python
from numpy import random

x = random.logistic(loc = 1, scale = 2, size = (2,3))
print(x)
                    
```

# Lojistik Daƒüƒ±lƒ±mƒ±n G√∂rselle≈ütirilmesi üß†

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.logistic(size = 1000), kind = "kde")
plt.show()
```

# Lojistik ve Normal Daƒüƒ±lƒ±m Arasƒ±ndaki Farklƒ±lƒ±klar üß†

‚ÑπÔ∏è *ƒ∞ki daƒüƒ±lƒ±m da hemen hemen aynƒ±dƒ±r.Fakat Lojistik Daƒüƒ±lƒ±m u√ß kƒ±sƒ±mlarda daha fazla alana sahiptir.Bu ≈üu demek: Ortalamadan uzak bir olayƒ±n meydana gelme olasƒ±lƒ±ƒüƒ±nƒ± daha fazla temsil eder.*<br>

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(scale=2, size=1000),
  "logistic": random.logistic(size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```
