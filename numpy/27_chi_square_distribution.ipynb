# Chi Square Distribution (Chi-Kare Daƒüƒ±lƒ±mƒ±) üöÄ

‚ÑπÔ∏è *Chi-Kare Daƒüƒ±lƒ±mƒ± hipotez doprulamasƒ± amacƒ±yla kullanƒ±lƒ±r.*<br>
‚ÑπÔ∏è *2 parametre deƒüeri alƒ±r:*<br>
**df**: Serbestlik Derecesi.*<br>
**size**: D√∂nen dizinin ≈üekli.*

*Serbestlik derecesi 2 ve boyutu 2x3 olan bir chi-kare daƒüƒ±lƒ±m √∂rneklemi √ßizelim:*<br>

```python
from numpy import random
x = random.chisquare(df = 2, size = (2,3))
print(x)
```

## Chi-Square Daƒüƒ±lƒ±m Grafiƒüi

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.chisquare(df = 1, size = 1000), kind = "kde")
plt.show()
```
