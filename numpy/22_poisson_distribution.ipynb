# Poission Daƒüƒ±lƒ±mƒ± üöÄ

‚ÑπÔ∏è *Poission Daƒüƒ±lƒ±mƒ± ayrƒ±k bir daƒüƒ±lƒ±mdƒ±r.*<br>
‚ÑπÔ∏è *Belirli bir zaman dilimi i√ßerisinde bir olayƒ±n ka√ß defa ger√ßekle≈üebileceƒüini tahmin eder.√ñrneƒüin bir bir ki≈üi g√ºnde 2 √∂ƒü√ºn yemek yiyorsa bu ki≈üinin g√ºnde 3 √∂ƒü√ºn yeme olasƒ±lƒ±ƒüƒ± nedir?*<br>
‚ÑπÔ∏è *ƒ∞ki parametre alƒ±r:*<br>
    ***lam**:oran veya bilinen olay sayƒ±sƒ±*<br>
    ***size**: d√∂nd√ºr√ºlen dizinin boyutu*

*Olu≈üum 2 i√ßin 1x10'luk rastgele daƒüƒ±lƒ±m √ºretelim:*

```python
from numpy import random

x = random.poisson(lam = 2, size = 10)
print(x)
```

## Poission Daƒüƒ±lƒ±mƒ±nƒ±n G√∂rselle≈ütirilmesi üß†

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.poisson(lam = 2, size = 1000))
plt.show()
```

## Normal Daƒüƒ±lƒ±m ile Poisson Daƒüƒ±lƒ±mƒ± Arasƒ±ndaki Fark üß†

‚ÑπÔ∏è *Binom Daƒüƒ±lƒ±mƒ±nda olduƒüu gibi burda da Normal Daƒüƒ±lƒ±m s√ºrekli iken Poisson Daƒüƒ±lƒ±mƒ± ayrƒ±ktƒ±r.<br>*
‚ÑπÔ∏è *Ancak yeterince b√ºy√ºk bir Poisson Daƒüƒ±lƒ±mƒ± i√ßin Binom Daƒüƒ±lƒ±mƒ±nda olduƒüu gibi belirli bir standart sapma ve ortalamaya sahip normal daƒüƒ±lƒ±ma benzeyeceƒüini g√∂rebiliriz.*

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {

    "normal" : random.normal(loc = 50, scale = 7, size = 1000),
    "poisson" : random.poisson(lam = 50, size = 1000)
}

sns.displot(data, kind = "kde")
plt.show()
```

## Binom Daƒüƒ±lƒ±mƒ± ile Poisson Daƒüƒ±lƒ±mƒ± Arasƒ±ndaki Fark üß†

‚ÑπÔ∏è *Binom Daƒüƒ±lƒ±mƒ± en fazla 2 olasƒ± sonuca sahiptir.Fakat Poisson Daƒüƒ±lƒ±mƒ±nda sƒ±nƒ±rsƒ±z sayƒ±da olasƒ± deƒüer elde edebiliriz*<br>
‚ÑπÔ∏è *Fakat √ßok b√ºy√ºk **n** ve sƒ±fƒ±ra yakƒ±n **p** deƒüerlerinde,**nxp** deƒüerinin **lam** deƒüerine neredeyse e≈üit olduƒüu ≈üekilde binom daƒüƒ±lƒ±mƒ± kimlik olarak possion daƒüƒ±lƒ±mƒ±na benzer.*

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sls

data = {

    "binom" : random.binomial( n = 1000, p = 0.01, size = 1000),
    "poisson" : random.poisson(lam = 10, size = 1000)
}

sls.displot(data, kind = "kde")
plt.show()
```
