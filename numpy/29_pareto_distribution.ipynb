# Pareto Distribution (Pareto Daƒüƒ±lƒ±mƒ±) üöÄ

‚ÑπÔ∏è *ƒ∞smini Pareto Yasasƒ±'ndan alan daƒüƒ±lƒ±m(%20'lik fakt√∂rler %80'lik daƒüƒ±lƒ±mƒ± olu≈üturur).*<br>
‚ÑπÔ∏è *ƒ∞ki parametre alƒ±r:*<br>
**a**: ≈ûekil parametresi.*<br>
**size**: D√∂nen dizinin boyutu.*

*2x3 boyutunda ≈üekli 2 olan Pareto Daƒüƒ±lƒ±m √∂rneklemi √ßizelim:*<br>

```python
from numpy import random

x = random.pareto(a = 2, size = (2, 3))
print(x)
```

## Pareto Daƒüƒ±lƒ±m Grafiƒüi

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.pareto(a=2, size=1000))

plt.show()
```
