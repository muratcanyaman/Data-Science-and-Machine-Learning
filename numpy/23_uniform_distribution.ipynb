‚ÑπÔ∏è *Her olayƒ±n ger√ßekle≈üme ihtimalinin e≈üit olduƒüu olasƒ±lƒ±klarƒ± ifade ederken kullanƒ±lan bir daƒüƒ±lƒ±mdƒ±r.*<br>
‚ÑπÔ∏è *√ñrneƒüin rastgele sayƒ±larƒ±n √ºretimi.*<br>
‚ÑπÔ∏è *3 parametre alƒ±r :*<br>
***low**: Alt sƒ±nƒ±r.Varsayƒ±lan olarak 0.0*<br>
***high**: √úst sƒ±nƒ±r.Varsayƒ±lan olarak 1.0*<br>
***size**: Geriye d√∂nd√ºr√ºlen dizinin boyutu*

```python
# 2x3 boyutunda Uniform daƒüƒ±lƒ±m √∂rneƒüi olu≈üturalƒ±m
from numpy import random
x = random.uniform(size = (2,3))
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.uniform(size = 1000), kind = 'kde')
plt.show()
```
