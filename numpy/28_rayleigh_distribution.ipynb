# Rayleigh Distribution(Rayleigh Daƒüƒ±lƒ±mƒ±) üöÄ

‚ÑπÔ∏è *Rayleigh Daƒüƒ±lƒ±mƒ± sinyal i≈ülemede kullanƒ±lƒ±r.*<br>
‚ÑπÔ∏è *2 parametre alƒ±r:*<br>
**scale**: Standart Sapma. Daƒüƒ±lƒ±mƒ±n ne kadar d√ºz olacaƒüƒ±nƒ± belirler.Varsayƒ±lan olarak 1'dir.*<br>
**size**: Geriye d√∂nd√ºr√ºlen dizinin ≈üekli.*<br>

*scale deƒüeri 2 ve boyutu 2x3 olan bir Rayleigh daƒüƒ±lƒ±m √∂rneklemi olu≈üturalƒ±m:*<br>

```python
from numpy import random

x = random.rayleigh(scale = 2, size = (2, 3))
print(x)
```

## Rayleigh Daƒüƒ±lƒ±m Grafiƒüi üß†

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.rayleigh(size = 1000), kind = "kde")
plt.show()
```
