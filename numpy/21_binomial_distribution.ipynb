# Binom Daƒüƒ±lƒ±mƒ±(Binomial Distribution) üöÄ

‚ÑπÔ∏è *Binom Daƒüƒ±lƒ±mƒ± bir Ayrƒ±k Daƒüƒ±lƒ±mdƒ±r.*<br>
‚ÑπÔ∏è *ƒ∞kili senaryolarƒ±n sonu√ßlarƒ±nƒ± tanƒ±mlar.√ñrneƒüin yazƒ± tura oyunu.Ya yazƒ± √ßƒ±kacaktƒ±r ya da tura.*<br>
‚ÑπÔ∏è *3 parametre alƒ±r*:<br>
*n - deneme sayƒ±sƒ±*<br>
*p - her bir denemenin ger√ßekle≈üme olasƒ±lƒ±ƒüƒ±*<br>
*size - geriye d√∂nd√ºr√ºlecek dizinin boyutu*

‚ö†Ô∏è ***Ayrƒ±k Daƒüƒ±lƒ±m**:Daƒüƒ±lƒ±m,ayrƒ±k olaylar k√ºmesinde tanƒ±mlanƒ±r.√ñrneƒüin yazƒ± tura atmanƒ±n sonucu sadece yazƒ± veya tura olacaƒüƒ±ndan bu bir ayrƒ±k daƒüƒ±lƒ±mdƒ±r.Oysa insanlarƒ±n boyu 170,170.1,170.11 olabileceƒüinden bu bir s√ºrekli bir daƒüƒ±lƒ±mdƒ±r.*<br>

yazƒ± tura i√ßin 10 deneme verilsin ve 10 farklƒ± veri noktasƒ± olu≈üturalƒ±m:*

```python
from numpy import random
x = random.binomial(n = 10, p = 0.5, size = 10)
print(x)
```

## Binom Daƒüƒ±lƒ±mƒ±nƒ±n G√∂rselle≈ütirilmesi üß†

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n = 10, p = 0.5, size = 1000))
plt.show()
```

## Normal Daƒüƒ±lƒ±m ile Binom Daƒüƒ±lƒ±mƒ± Arasƒ±ndaki Fark üß†

‚ÑπÔ∏è *Normal Daƒüƒ±lƒ±m ile Binom Daƒüƒ±lƒ±mƒ± arasƒ±ndaki ana fark, normal daƒüƒ±lƒ±m s√ºrekli iken binom daƒüƒ±lƒ±mƒ± ayrƒ±ktƒ±r.Fakat yeterince veri noktasƒ± ile belirli konum ve √∂l√ßekte normal daƒüƒ±lƒ±ma benzeyebilir.*

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "normal" : random.normal( loc = 50, scale = 5, size = 1000),
    "binomial" : random.binomial( n = 100, p = 0.5, size =  1000)
}

sns.displot(data, kind = "kde")
plt.show()
```
