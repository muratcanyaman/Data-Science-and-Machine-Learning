## Var Olan Bir Dizinin Veri Tipini DÃ¶nÃ¼ÅŸtÃ¼rme ğŸ§ 	

â„¹ï¸	*Var olan bir dizinin veri tipini deÄŸiÅŸtirmenin en iyi yolu, dizinin bir kopyasÄ±nÄ± oluÅŸturan **astype()** metodunu kullanmaktÄ±r.*<br>
â„¹ï¸	***astype()** fonksiyonu dizinin bir kopyasÄ±nÄ± oluÅŸturur,ve veri tipini bir parametre olarak girmenize olanak saÄŸlar*<br>
â„¹ï¸	*Veri tipini belirtirken; float iÃ§in **f**, integer iÃ§in **i** vb. kullanabileceÄŸiniz gibi direkt olarak **float**, **integer** ÅŸeklinde de parametre girebilirsiniz.*

```python
# float veri tipinden integer veri tipine dÃ¶nÃ¼ÅŸÃ¼m
import numpy as np
arr = np.array([3.14, 7.81, 11.13])

new_arr = arr.astype('i')
print(new_arr)
print(new_arr.dtype)
```

*AynÄ± iÅŸlemi **int** parametresi kullanarak yapalÄ±m:*<br>

```python
import numpy as np
arr = np.array([3.14, 7.81, 11.13])

new_arr = arr.astype(int)
print(new_arr)
print(new_arr.dtype)
```

*Veri tipini integer'dan boolean'a dÃ¶nÃ¼ÅŸtÃ¼rme:*

```python
import numpy as np
arr = np.array([0,1,2,3])

new_arr = arr.astype(bool)
print(new_arr)
print(new_arr.dtype)
```
