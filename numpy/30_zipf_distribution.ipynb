# Zipf Distribution (Zipf Daƒüƒ±lƒ±mƒ±) üöÄ

‚ÑπÔ∏è *Zipf Daƒüƒ±lƒ±mƒ±,Zipf Yasasƒ±'na g√∂re verileri √∂rneklemek i√ßin kullanƒ±lƒ±r.*<br>

üìå **Zipf Kanunu**: Bir koleksiyonda, en yaygƒ±n n'inci terim en yaygƒ±n terimin 1/n katƒ±dƒ±r. √ñrneƒüin ƒ∞ngilizce'de en yaygƒ±n 5.kelime,en yaygƒ±n kelimenin 1/5'i sƒ±klƒ±kta ge√ßer.*<br>

‚ÑπÔ∏è *2 parametre alƒ±r:*<br>
**a**: Daƒüƒ±lƒ±m parametresi*<br>
**size**: D√∂nen dizinin ≈üekli*

*Daƒüƒ±lƒ±m parametresi 2 ve boyutu 2x3 olan zipf daƒüƒ±lƒ±m √∂rneklemi √ßizelim:*<br>

```python
from numpy import random

x = random.zipf(a = 2, size = (2, 3))
print(x)
```

## Zipf Daƒüƒ±lƒ±m Grafiƒüi

*1000 puan fakat daha anlamlƒ± grafik √ßizebilmek i√ßin deƒüeri 10'dan k√º√ß√ºk olanlarƒ± alalƒ±m:*<br>

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.zipf(a=2, size=1000)
sns.displot(x[x<10])

plt.show()
```
